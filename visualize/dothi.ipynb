```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# ng dn ti tp JSON
file_path = '../dothi.json'  # Thay th bng ng dn chnh xc ti tp ca bn

# c d liu t tp JSON
with open(file_path, 'r', encoding='utf-8') as file:
    data = json.load(file)

# Chuyn i d liu thnh DataFrame
df = pd.DataFrame(data)

# Lc ra cc qun v huyn c tn bt u bng "Qun" hoc "Huyn"
filtered_df = df[df['Qun'].str.startswith(('Qun', 'Huyn'))]

# m s lng cn h theo tng qun  lc
district_counts = filtered_df['Qun'].value_counts()

# To biu  ct
# iu chnh kch thc ca biu 
plt.figure(figsize=(12, 8))

# To biu  ct v xoay nhn trc x
district_counts.plot(kind='bar', color='skyblue')
plt.title('S lng cn h theo qun', fontsize=14)
plt.xlabel('Qun', fontsize=12)
plt.ylabel('S lng cn h', fontsize=12)
plt.xticks(rotation=90)  # Xoay nhn 90 
plt.tight_layout()

# Hin th biu 
plt.show()
```
