```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# Đường dẫn tới tệp JSON
file_path = '../dothi.json'  # Thay thế bằng đường dẫn chính xác tới tệp của bạn

# Đọc dữ liệu từ tệp JSON
with open(file_path, 'r', encoding='utf-8') as file:
    data = json.load(file)

# Chuyển đổi dữ liệu thành DataFrame
df = pd.DataFrame(data)

# Lọc ra các quận và huyện có tên bắt đầu bằng "Quận" hoặc "Huyện"
filtered_df = df[df['Quận'].str.startswith(('Quận', 'Huyện'))]

# Đếm số lượng căn hộ theo từng quận đã lọc
district_counts = filtered_df['Quận'].value_counts()

# Tạo biểu đồ cột
# Điều chỉnh kích thước của biểu đồ
plt.figure(figsize=(12, 8))

# Tạo biểu đồ cột và xoay nhãn trục x
district_counts.plot(kind='bar', color='skyblue')
plt.title('Số lượng căn hộ theo quận', fontsize=14)
plt.xlabel('Quận', fontsize=12)
plt.ylabel('Số lượng căn hộ', fontsize=12)
plt.xticks(rotation=90)  # Xoay nhãn 90 độ
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```
