NORMALIZE LEGAL STATUS ATTRIBUTE

0: No title (temporary documents)
	•	Values such as “Sales contract (HDMB)”, “Purchase agreement”, “Waiting for the title”, “Lease contract”.

1: Title available (red/ pink book)
	•	Values such as “Red book”, “Pink book”, “Red/ Pink book”, “Red book with official ownership”, “Long-term red book”, “Separate red book”, “Red book ready for transaction”.

2: Special title (limited or conditional)
	•	Values such as “50-year title”, “Long-term title + 50-year purchase agreement (HDMB)”.

3: Full legal documentation, safe but unclear title type
	•	Values such as “Complete legal documentation, ensuring secure real estate transactions”, “Fully prepared for transactions”.


 NORMALIZE FURNITURE ATTRIBUTE
 

	•	0: Unfurnished: Descriptions such as “Unfurnished”.
	•	1: Basic furnishings: Descriptions such as “Basic”, “Basic furnishings completed by the investor”, “Basic - Newly built and handed over by the developer”.
	•	2: Fully furnished: Descriptions such as “Fully furnished”, “Complete furnishings, ready for immediate move-in”.
	•	3: High-end furnishings: Descriptions such as “High-end furnishings”, “Full high-end furnishings”, “Fully equipped with built-in high-end furnishings imported from Europe”.
	•	4: Luxury furnishings: Descriptions such as “Full luxury European imported furnishings”, “Fully furnished with high-end items from Smegg, Villory&Boss”.


Circular Encoding là mã hoá với sin và cos để biểu diễn các hướng. Mỗi hướng sẽ được biểu diễn bằng một cặp giá trị (x, y) trên mặt phẳng.
Lý do chọn Circular Encoding:

	•	Giữ được tính tuần hoàn của dữ liệu: Giúp mô hình không hiểu nhầm các hướng gần nhau là các giá trị cách xa nhau (như trong Label Encoding).
	•	Tránh thứ tự giả: Không tạo ra thứ tự giữa các hướng như Label Encoding, đồng thời thể hiện sự liên kết giữa các hướng một cách tự nhiên.

```python
import numpy as np
import pandas as pd
directions = ['None', 'Đông', 'Tây', 'Nam', 'Bắc', 'Đông - Nam', 'Tây - Nam', 'Đông - Bắc', 'Tây - Bắc']
N = len(directions) - 1  
def circular_encoding(index, N):
    angle = 2 * np.pi * index / N
    x = np.cos(angle)
    y = np.sin(angle)
    return x, y
encoded_data = pd.DataFrame(columns=['Hướng', 'x', 'y'])

for i, direction in enumerate(directions):
    if direction == 'None':
        x, y = 0, 0  
    else:
        x, y = circular_encoding(i - 1, N)  
    new_row = pd.DataFrame({'Hướng': [direction], 'x': [x], 'y': [y]})
    encoded_data = pd.concat([encoded_data, new_row], ignore_index=True)
print(encoded_data)
```
